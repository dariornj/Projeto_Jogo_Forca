```python
# Organizando as 'imagens' de erro em uma lista

erro_0 = """
   |-------
   |      |
   |    
   |    
   |    
   |     
   |     
___|___ 
"""
erro_1 = """
   |-------
   |      |
   |      _
   |     |_|
   |      
   |     
   |     
___|___ 
"""

erro_2 = ("""
   |-------
   |      |
   |      _
   |     |_|
   |      |
   |      |
   |     
___|___ 
""")

erro_3 = ("""
   |-------
   |      |
   |      _
   |     |_|
   |    --|
   |      |
   |     
___|___ 
""")

erro_4 = ("""
   |-------
   |      |
   |      _
   |     |_|
   |    --|--
   |      |
   |     
___|___ 
""")

erro_5 = ("""
   |-------
   |      |
   |      _
   |     |_|
   |    --|--
   |      |
   |     / 
___|___ 
""")

erro_6 = ("""
   |-------
   |      |
   |      _
   |     |_|
   |    --|--
   |      |
   |     / \\
___|___ 
""")

erros = [erro_0, erro_1, erro_2, erro_3, erro_4, erro_5, erro_6]
```

```python
# Abrindo o arquivo .txt com os nomes das frutas como uma lista

with open ('frutas.txt', 'r', encoding = 'utf-8') as frutas:
    lista_frutas = []
    for fruta in frutas:
        lista_frutas.append(fruta.strip())  # Remoção do \n
    lista_frutas.pop(-1)   # Remoção da linha vazia
```

```python
# Funções utilizadas no programa

def escolha_fruta(lista_frutas):
    """ Função para escolher um elemento aleatório numa lista """
    import random
    n_frutas = len(lista_frutas)
    fruta_escolhida = lista_frutas[random.randint(0, n_frutas)]
    return fruta_escolhida

def string_impressao(string):
    tamanho = len(string)
    underline = ' '*(tamanho)

    palavra_separada = ''.join (i+j for i, j, in zip(string, underline))
    return palavra_separada

def subs_letra (fruta_escolhida, fruta_jogador, letra_escolhida):
    """Função para substituir a letra da forca de acordo com a letra escolhida pelo usuário"""

    for idx in range(len(fruta_escolhida)):
        fruta_jogador_list = list(fruta_jogador)
        if fruta_escolhida[idx] == letra_escolhida:
            fruta_jogador_list[idx] = letra_escolhida
        fruta_jogador = ''.join(fruta_jogador_list)
        
    return fruta_jogador
```

```python
def iniciar_jogo():
    fruta_escolhida = escolha_fruta(lista_frutas)
    jogando = True
    letras_escolhidas = []
    erro = 0
    n_erros = 6
    fruta_jogador = '_'*len(fruta_escolhida)

    #Impressão de tela inicial:
    print('Bem vindo ao jogo de forca!')
    print(erros[erro])
    print(string_impressao(fruta_jogador))

    while jogando:
        if fruta_jogador != fruta_escolhida:
            
            while erro < n_erros:
                letra_escolhida = input('Digite a letra')

                # Verificando se a letra já foi escolhida anteriormente
                if letra_escolhida in letras_escolhidas:
                    print('Letra já escolhida, tente novamente.')
                    break
                else:
                    letras_escolhidas.append(letra_escolhida)

                # Verificando se a letra existe na fruta escolhida e realizando a substituição
                fruta_jogador_anterior = fruta_jogador           
                fruta_jogador = subs_letra (fruta_escolhida, fruta_jogador_anterior, letra_escolhida)

                if fruta_jogador == fruta_jogador_anterior:
                    erro +=1
                    print(f'Letra não encontrada na fruta escolhida. Você possui mais {6 - erro} erros possíveis')
                    if erro == 6:
                        jogando = False
                        print(erros[erro])
                        print(f'A fruta é {fruta_escolhida}. Inicie novo jogo.')
                        break

                    print(erros[erro])
                    print(string_impressao(fruta_jogador))
                else:
                    if fruta_jogador == fruta_escolhida:
                        jogando = False
                        print(f'Parabéns, você acertou! A palavra é {fruta_escolhida} e você teve {erro} erros.')
                        break
                    else:
                        print(f'Ótimo! Escolha outra letra')
                        print(erros[erro])
                        print(string_impressao(fruta_jogador))
```

```python
# Iniciando o jogo

iniciar_jogo()
```
